```python
from tkinter import *

# T·∫°o c·ª≠a s·ªï
root = Tk()
root.title("Student Information")
root.geometry("350x200")

def check_filled(*args):
    if name_var.get() and id_var.get() and pass_var.get():
        submit_btn.pack(pady=10)
    else:
        submit_btn.pack_forget()

name_var = StringVar()
id_var = StringVar()
pass_var = StringVar()

name_var.trace_add("write", check_filled)
id_var.trace_add("write", check_filled)
pass_var.trace_add("write", check_filled)

# Label + Entry
Label(root, text="Student Name:").pack(anchor="w", padx=20)
Entry(root, textvariable=name_var).pack(fill="x", padx=20)

Label(root, text="Student ID:").pack(anchor="w", padx=20, pady=5)
Entry(root, textvariable=id_var).pack(fill="x", padx=20)

Label(root, text="Password:").pack(anchor="w", padx=20, pady=5)
Entry(root, textvariable=pass_var, show="*").pack(fill="x", padx=20)

# N√∫t Submit (ban ƒë·∫ßu ·∫©n)
submit_btn = Button(root, text="Submit")

# Ch·∫°y ch∆∞∆°ng tr√¨nh
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

# H√†m t√≠nh GCD (kh√¥ng d√πng math.gcd)
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

# H√†m x·ª≠ l√Ω khi ch·ªçn ch·ª©c nƒÉng
def calculate():
    try:
        m = int(entry_m.get())
        n = int(entry_n.get())
        func = combo.get()

        if func == "GCD":
            result = gcd(m, n)
            result_label.config(text=f"GCD(m, n) = {result}")
        else:
            result = (m * n) // gcd(m, n)
            result_label.config(text=f"LCM(m, n) = {result}")
    except:
        result_label.config(text="Please enter valid integers")

# T·∫°o c·ª≠a s·ªï
root = tk.Tk()
root.title("tk")
root.geometry("500x250")

# Nh·∫≠p m
tk.Label(root, text="Enter value of m:").pack(pady=5)
entry_m = tk.Entry(root)
entry_m.pack()

# Nh·∫≠p n
tk.Label(root, text="Enter value of n:").pack(pady=5)
entry_n = tk.Entry(root)
entry_n.pack()

# Ch·ªçn ch·ª©c nƒÉng
tk.Label(root, text="Choose function:").pack(pady=5)
combo = ttk.Combobox(root, values=["GCD", "LCM"])
combo.current(0)
combo.pack()

# N√∫t Validate
tk.Button(root, text="Validate", command=calculate).pack(pady=10)

# Hi·ªÉn th·ªã k·∫øt qu·∫£
result_label = tk.Label(root, text="")
result_label.pack(pady=10)

# Ch·∫°y ch∆∞∆°ng tr√¨nh
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

zodiacs = [
    "T√Ω ", "S·ª≠u ", "D·∫ßn ", "M√£o ",
    "Th√¨n ", "T·ªµ", "Ng·ªç üê¥", "M√πi ",
    "Th√¢n ", "D·∫≠u ", "Tu·∫•t ", "H·ª£i "
]

def convert_year():
    global entry_year, result_label
    try:
        year = int(entry_year.get())
        zodiac = zodiacs[(year - 4) % 12]

        messagebox.showinfo(
            "K·∫øt qu·∫£",
            f"NƒÉm √¢m l·ªãch: {zodiac}"
        )

        result_label.config(
            text=f"NƒÉm {year}\nCon gi√°p: {zodiac}",
            font=("Arial", 18)
        )

    except:
        messagebox.showerror("L·ªói", "Vui l√≤ng nh·∫≠p nƒÉm h·ª£p l·ªá")

root = tk.Tk()
root.title("Chuy·ªÉn ƒë·ªïi nƒÉm √¢m l·ªãch")
root.geometry("400x300")

tk.Label(root, text="Nh·∫≠p nƒÉm d∆∞∆°ng l·ªãch:", font=("Arial", 12)).pack(pady=10)

entry_year = tk.Entry(root, font=("Arial", 12))
entry_year.pack()

tk.Button(root, text="Chuy·ªÉn ƒë·ªïi", command=convert_year).pack(pady=15)

result_label = tk.Label(root, text="", fg="blue")
result_label.pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def calculate_bmi():
    global entry_weight, entry_height, result_label, status_label
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())

        if height <= 0 or weight <= 0:
            raise ValueError

        bmi = weight / (height ** 2)

        # Ph√¢n lo·∫°i BMI
        if bmi < 18.5:
            status = "G·∫ßy"
        elif bmi < 25:
            status = "B√¨nh th∆∞·ªùng"
        elif bmi < 30:
            status = "Th·ª´a c√¢n"
        else:
            status = "B√©o ph√¨"

        result_label.config(text=f"BMI = {bmi:.2f}")
        status_label.config(text=f"T√¨nh tr·∫°ng: {status}")

    except:
        messagebox.showerror("L·ªói", "Vui l√≤ng nh·∫≠p c√¢n n·∫∑ng v√† chi·ªÅu cao h·ª£p l·ªá")

root = tk.Tk()
root.title("T√≠nh ch·ªâ s·ªë BMI")
root.geometry("400x300")

tk.Label(root, text="C√¢n n·∫∑ng (kg):").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chi·ªÅu cao (m):").pack(pady=5)
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="T√≠nh BMI", command=calculate_bmi).pack(pady=15)

result_label = tk.Label(root, text="", font=("Arial", 14), fg="blue")
result_label.pack(pady=10)

status_label = tk.Label(root, text="", font=("Arial", 12))
status_label.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show_exercise(title, content):
    messagebox.showinfo(title, content)

def create_chapter_menu(menu_bar, chapter_number):
    chapter_menu = tk.Menu(menu_bar, tearoff=0)

    for i in range(1, 10):
        if chapter_number == 1 and i == 1:
            chapter_menu.add_command(
                label=f"{chapter_number}.{i}",
                command=lambda: show_exercise(
                    "L·∫≠p tr√¨nh h∆∞·ªõng ƒë·ªëi t∆∞·ª£ng",
                    "Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c hai"
                )
            )
        else:
            chapter_menu.add_command(
                label=f"{chapter_number}.{i}",
                command=lambda m=chapter_number, n=i: show_exercise(
                    "B√†i t·∫≠p",
                    f"B√†i t·∫≠p {m}.{n}"
                )
            )

    chapter_menu.add_separator()
    chapter_menu.add_command(label="Exit", command=root.quit)

    menu_bar.add_cascade(label=f"Ch∆∞∆°ng {chapter_number}", menu=chapter_menu)

root = tk.Tk()
root.title("H·ªá th·ªëng b√†i t·∫≠p n√¢ng cao Python")
root.geometry("600x400")

menu_bar = tk.Menu(root)
root.config(menu=menu_bar)

for chapter in range(1, 8):
    create_chapter_menu(menu_bar, chapter)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox

def calculate_fare():
    global entry_km, entry_wait, car_type, result_label
    try:
        km = float(entry_km.get())
        wait = int(entry_wait.get())
        car = car_type.get()

        # Ti·ªÅn m·ªü c·ª≠a
        total = 12000

        # T√≠nh ti·ªÅn qu√£ng ƒë∆∞·ªùng
        if car == "4 ch·ªó":
            if km <= 0.8:
                distance_cost = 0
            elif km <= 30:
                distance_cost = (km - 0.8) * 15300
            else:
                distance_cost = (30 - 0.8) * 15300 + (km - 30) * 12100
        else:  # 7 ch·ªó
            if km <= 0.8:
                distance_cost = 0
            elif km <= 30:
                distance_cost = (km - 0.8) * 16100
            else:
                distance_cost = (30 - 0.8) * 16100 + (km - 30) * 13800

        # T√≠nh ti·ªÅn ch·ªù
        if wait <= 5:
            wait_cost = 0
        else:
            wait_cost = (wait - 5) * 750

        total += distance_cost + wait_cost

        result_label.config(
            text=f"T·ªïng ti·ªÅn taxi: {int(total):,} ƒë·ªìng"
        )

    except:
        messagebox.showerror("L·ªói", "Vui l√≤ng nh·∫≠p d·ªØ li·ªáu h·ª£p l·ªá")

root = tk.Tk()
root.title("T√≠nh c∆∞·ªõc taxi")
root.geometry("450x350")

tk.Label(root, text="Qu√£ng ƒë∆∞·ªùng (km):").pack(pady=5)
entry_km = tk.Entry(root)
entry_km.pack()

tk.Label(root, text="Th·ªùi gian ch·ªù (ph√∫t):").pack(pady=5)
entry_wait = tk.Entry(root)
entry_wait.pack()

tk.Label(root, text="Lo·∫°i xe:").pack(pady=5)
car_type = ttk.Combobox(root, values=["4 ch·ªó", "7 ch·ªó"])
car_type.current(0)
car_type.pack()

tk.Button(root, text="T√≠nh ti·ªÅn", command=calculate_fare).pack(pady=15)

result_label = tk.Label(root, text="", font=("Arial", 14), fg="blue")
result_label.pack(pady=15)

root.mainloop()
```
