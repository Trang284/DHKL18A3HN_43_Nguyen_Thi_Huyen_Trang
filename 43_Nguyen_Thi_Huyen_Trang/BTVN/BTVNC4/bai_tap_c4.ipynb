```python
import pandas as pd
import numpy as np

mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

s_from_list = pd.Series(mylist)
print("Series từ list:")
print(s_from_list)

s_from_array = pd.Series(myarr)
print("\nSeries từ numpy array:")
print(s_from_array)

s_from_dict = pd.Series(mydict)
print("\nSeries từ dictionary:")
print(s_from_dict)
```

```python
import pandas as pd

euro12 = pd.read_csv("euro12.csv")

# In thông tin cơ bản
print("Type:", type(euro12))
print("Shape:", euro12.shape)
print("Columns:", euro12.columns.tolist())

# 1. In giá trị cột Goals
print("\nCột Goals:")
print(euro12['Goals'])

# 2. Số đội tham gia Euro 2012
num_teams = euro12['Team'].nunique()
print("\nSố đội tham gia Euro 2012:", num_teams)

# 3. Thông tin của Euro 2012
print("\nThông tin Euro 2012:")
print(euro12.info())

# 4. Tạo DataFrame discipline với các cột Team, Yellow Cards, Red Cards
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\nDataFrame discipline:")
print(discipline)

# 5. Sắp xếp discipline theo Red Cards và Yellow Cards giảm dần
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\nDiscipline sắp xếp giảm dần:")
print(discipline_sorted)

# 6. Trung bình Yellow Cards
yellow_avg = euro12['Yellow Cards'].mean()
print("\nTrung bình Yellow Cards:", yellow_avg)

teams_goals_gt6 = euro12[euro12['Goals'] > 6]
print("\nCác đội ghi hơn 6 bàn thắng:")
print(teams_goals_gt6)

teams_G = euro12[euro12['Team'].str.startswith('G')]
print("\nCác đội tên bắt đầu bằng 'G':")
print(teams_G)

print("\n7 cột đầu của euro12:")
print(euro12.iloc[:, :7])

print("\nTất cả cột trừ 3 cột cuối:")
print(euro12.iloc[:, :-3])

print("\nCác cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards:")
print(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])

teams_selected = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]
print("\nTeam và Shooting Accuracy của England, Italy, Russia:")
print(teams_selected[['Team', 'Shooting Accuracy']])
```

```python
import pandas as pd

drink = pd.read_csv("drinks.csv", index_col=0)
# Thông tin cơ bản
print("Type:", type(drink))
print("Shape:", drink.shape)
print("Columns:", drink.columns.tolist())
print("\n5 dòng đầu:")
print(drink.head())
print("\n5 dòng cuối:")
print(drink.tail())

beer_mean_by_continent = drink.groupby('continent')['beer_servings'].mean()
print("\nSố lượng bia tiêu thụ trung bình theo châu lục:")
print(beer_mean_by_continent)

wine_stats_by_continent = drink.groupby('continent')['wine_servings'].describe()
print("\nThông tin thống kê rượu vang theo châu lục:")
print(wine_stats_by_continent)

mean_beer_spirit = drink.groupby('continent')[['beer_servings','spirit_servings']].mean()
print("\nTrung bình bia và rượu theo châu lục:")
print(mean_beer_spirit)

median_beer_spirit = drink.groupby('continent')[['beer_servings','spirit_servings']].median()
print("\nTrung vị bia và rượu theo châu lục:")
print(median_beer_spirit)

spirit_stats = drink.groupby('continent')['spirit_servings'].agg(['mean','max','min'])
print("\nSpirit servings theo châu lục:")
print(spirit_stats)

sorted_drink = drink.sort_values('beer_servings')
print("\n5 quốc gia tiêu thụ bia ít nhất:")
print(sorted_drink.head())
print("\n5 quốc gia tiêu thụ bia nhiều nhất:")
print(sorted_drink.tail())
```
