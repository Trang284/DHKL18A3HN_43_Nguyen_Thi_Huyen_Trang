```python
import sqlite3

# Tạo và kết nối tới cơ sở dữ liệu SQLite (tạo mới nếu chưa có)
conn = sqlite3.connect("example.db")

# Tạo con trỏ
cursor = conn.cursor()

# Lấy version của SQLite
cursor.execute("SELECT sqlite_version();")

# In ra version
version = cursor.fetchone()
print("SQLite version:", version[0])

# Đóng kết nối
conn.close()
```

```python
import sqlite3

# Kết nối tới cơ sở dữ liệu SQLite đã có
conn = sqlite3.connect("student.db")

# Tạo con trỏ
cursor = conn.cursor()

# Truy vấn liệt kê các bảng
cursor.execute("""
SELECT name FROM sqlite_master
WHERE type='table';
""")

# In ra danh sách các bảng
tables = cursor.fetchall()
print("Danh sách các bảng trong CSDL:")
for table in tables:
    print(table[0])

# Đóng kết nối
conn.close()
```

Bài 5.5:

```python
import sqlite3
import pandas as pd

# Kết nối SQLite
conn = sqlite3.connect("student.db")

# Tạo DataFrame dữ liệu mẫu
df = pd.DataFrame({
    "name": ["An", "Binh", "Chi"],
    "age": [20, 21, 22]
})

# Ghi DataFrame vào bảng SQLite
df.to_sql(
    name="Student",
    con=conn,
    if_exists="replace",
    index=False
)

# Đọc tất cả dữ liệu từ bảng
result = pd.read_sql("SELECT * FROM Student", conn)

# Hiển thị các bản ghi
print(result)

# Đóng kết nối
conn.close()
```

```python
import sqlite3
import pandas as pd

# Kết nối tới CSDL SQLite
conn = sqlite3.connect("student.db")

# Đọc dữ liệu từ bảng
df = pd.read_sql("SELECT * FROM Student", conn)

# Cập nhật toàn bộ cột age
df["age"] = df["age"] + 1

# Ghi lại DataFrame vào bảng SQLite
df.to_sql(
    name="Student",
    con=conn,
    if_exists="replace",
    index=False
)

print("Đã cập nhật toàn bộ cột age trong bảng Student")

# Đóng kết nối
conn.close()
```

```python
import sqlite3

def delete_student_by_name(db_name, student_name):
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    cursor.execute(
        "DELETE FROM Student WHERE name = ?",
        (student_name,)
    )

    conn.commit()
    conn.close()

# Gọi hàm
delete_student_by_name("student.db", "An")
```

Bài 5.10:

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("product.db")

df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Pen", "Book", "Bag"],
    "price": [10.5, 25.0, 120.0],
    "amount": [100, 50, 20]
})

df.to_sql("product", conn, if_exists="replace", index=False)

df = pd.read_sql("SELECT * FROM product", conn)
print(df)

new_product = pd.DataFrame({
    "id": [4],
    "name": ["Notebook"],
    "price": [45.0],
    "amount": [30]
})

df = pd.concat([df, new_product], ignore_index=True)
df.to_sql("product", conn, if_exists="replace", index=False)

result = df[df["name"].str.contains("Book")]
print(result)

df.loc[df["id"] == 2, ["price", "amount"]] = [30.0, 60]
df.to_sql("product", conn, if_exists="replace", index=False)

df = df[df["id"] != 1]
df.to_sql("product", conn, if_exists="replace", index=False)

sorted_df = df.sort_values(by="price")
print(sorted_df)

sorted_df.to_sql("product", conn, if_exists="replace", index=False)
conn.close()
```

Bài 5.11:

```python
import sqlite3

conn = sqlite3.connect("ql_nhan_vien.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS PHONG (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    price REAL NOT NULL,
    amount INTEGER NOT NULL
)
""")

conn.commit()

# HIỂN THỊ BẢNG PHONG
cursor.execute("SELECT id, name, price, amount FROM PHONG")
print("BẢNG PHONG:")
for row in cursor.fetchall():
    print(row)

phong_data = [
    (1, "Kế toán", 1000.0, 5),
    (2, "Nhân sự", 1200.0, 3),
    (3, "CNTT", 1500.0, 7)
]

cursor.executemany(
    "INSERT OR IGNORE INTO PHONG VALUES (?, ?, ?, ?)",
    phong_data
)

conn.commit()

cursor.execute("SELECT id, name, price, amount FROM PHONG")
print("\nSAU KHI THÊM PHÒNG:")
for row in cursor.fetchall():
    print(row)

cursor.execute("""
CREATE TABLE IF NOT EXISTS NHAN_VIEN (
    id INTEGER PRIMARY KEY,
    ho_ten TEXT NOT NULL,
    tuoi INTEGER NOT NULL,
    dia_chi TEXT NOT NULL,
    luong REAL NOT NULL,
    id_phong INTEGER,
    FOREIGN KEY (id_phong) REFERENCES PHONG(id)
)
""")

conn.commit()

print("\nĐÃ TẠO BẢNG NHAN_VIEN")

nhanvien_data = [
    (1, "Nguyễn Văn A", 25, "Đà Nẵng", 8000.0, 1),
    (2, "Trần Thị B", 30, "Huế", 9000.0, 2),
    (3, "Lê Văn C", 28, "Quảng Nam", 10000.0, 3)
]

cursor.executemany(
    "INSERT OR IGNORE INTO NHAN_VIEN VALUES (?, ?, ?, ?, ?, ?)",
    nhanvien_data
)

conn.commit()

cursor.execute("SELECT * FROM NHAN_VIEN")
print("\nBẢNG NHAN_VIEN:")
for row in cursor.fetchall():
    print(row)

cursor.execute("""
SELECT NHAN_VIEN.id, ho_ten, tuoi, dia_chi, luong, PHONG.name
FROM NHAN_VIEN
JOIN PHONG ON NHAN_VIEN.id_phong = PHONG.id
""")

print("\nDANH SÁCH NHÂN VIÊN + PHÒNG:")
for row in cursor.fetchall():
    print(row)

conn.close()
```
