```python
import numpy as np

np_baseball = np.loadtxt("baseball_2D.txt")
print("Bước 0: Đã đọc file baseball_2D.txt")

print("Bước 1:")
print("Kiểu dữ liệu:", np_baseball.dtype)
print("Kích thước:", np_baseball.shape)

print("\nBước 2: Dòng thứ 50:", np_baseball[49])

np_weight = np_baseball[:, 1]
print("\nBước 3: np_weight:", np_weight)

print("\nBước 4: Chiều cao vận động viên thứ 124:", np_baseball[123, 0])

avg_height = np.mean(np_baseball[:, 0])
avg_weight = np.mean(np_baseball[:, 1])
print("\nBước 5:")
print("Chiều cao trung bình:", round(avg_height, 2), "inch")
print("Cân nặng trung bình:", round(avg_weight, 2), "pound")

correlation = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("\nBước 6: Hệ số tương quan chiều cao - cân nặng:", round(correlation, 3))
if correlation > 0.5:
    print("Nhận xét: Có tương quan thuận mạnh giữa chiều cao và cân nặng.")
elif correlation > 0:
    print("Nhận xét: Có tương quan thuận yếu giữa chiều cao và cân nặng.")
elif correlation < -0.5:
    print("Nhận xét: Có tương quan nghịch mạnh giữa chiều cao và cân nặng.")
elif correlation < 0:
    print("Nhận xét: Có tương quan nghịch yếu giữa chiều cao và cân nặng.")
else:
    print("Nhận xét: Không có tương quan giữa chiều cao và cân nặng.")
```

Bài 3.8:

```python
import numpy as np

heights = [190, 175, 180, 185, 178, 192, 170, 177, 182, 176, 195, 168]
positions = ['GK', 'M', 'A', 'D', 'M', 'GK', 'D', 'A', 'M', 'D', 'GK', 'A']

# Lưu heights vào file heights.txt
with open("heights.txt", "w") as f:
    for h in heights:
        f.write(f"{h}\n")

# Lưu positions vào file positions.txt
with open("positions.txt", "w") as f:
    for p in positions:
        f.write(f"{p}\n")

print("Đã tạo file heights.txt và positions.txt với dữ liệu mẫu.")

with open("heights.txt", "r") as f:
    heights_list = [float(line.strip()) for line in f]

with open("positions.txt", "r") as f:
    positions_list = [line.strip() for line in f]

# Bước 1a: Tạo numpy array từ positions
np_positions = np.array(positions_list)
print("\nBước 1a: np_positions:", np_positions)
print("Kiểu dữ liệu np_positions:", np_positions.dtype)

# Bước 1b: Tạo numpy array từ heights
np_heights = np.array(heights_list, dtype=float)
print("\nBước 1b: np_heights:", np_heights)
print("Kiểu dữ liệu np_heights:", np_heights.dtype)

# Bước 2: Chiều cao trung bình
gk_mean_height = np_heights[np_positions == 'GK'].mean()
print("\nBước 2: Chiều cao trung bình GK:", round(gk_mean_height, 2))

# Bước 3: Chiều cao trung bình ngoại trừ GK
non_gk_mean_height = np_heights[np_positions != 'GK'].mean()
print("Bước 3: Chiều cao trung bình vị trí khác GK:", round(non_gk_mean_height, 2))

# Bước 4: Tạo mảng players với dtype phức hợp
players_dtype = np.dtype([('position', 'U5'), ('height', 'float')])
players = np.zeros(len(heights_list), dtype=players_dtype)
players['position'] = np_positions
players['height'] = np_heights
print("\nBước 4: Mảng players:\n", players)

# Bước 5: Sắp xếp players theo chiều cao
players_sorted = np.sort(players, order='height')
print("\nBước 5: Players sắp xếp theo chiều cao:\n", players_sorted)

# Vị trí cao nhất và thấp nhất
tallest = players_sorted[-1]
shortest = players_sorted[0]
print("\nVị trí cao nhất:", tallest['position'], "với chiều cao:", tallest['height'])
print("Vị trí thấp nhất:", shortest['position'], "với chiều cao:", shortest['height'])
```
