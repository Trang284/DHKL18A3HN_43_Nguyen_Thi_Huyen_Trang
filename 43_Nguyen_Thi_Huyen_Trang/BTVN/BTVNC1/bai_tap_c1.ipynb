##### Bài 2:

```python
class TS:
    def __init__(self, ho_ten, toan, ly, hoa):
        self.ho_ten = ho_ten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa
    def tong_diem(self):
        return self.toan + self.ly + self.hoa
    def in_thong_tin(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Toán: {self.toan}, Lý: {self.ly}, Hoá: {self.hoa}")
        print(f"Tổng điểm: {self.tong_diem()}")

# Nhập danh sách thí sinh
ds = []
n = int(input("Nhập số lượng thí sinh: "))
for i in range(n):
    print(f"--- Thí sinh thứ {i+1} ---")
    ho_ten = input("Họ tên: ")
    toan = float(input("Điểm Toán: "))
    ly = float(input("Điểm Lý: "))
    hoa = float(input("Điểm Hoá: "))
    ts = TS(ho_ten, toan, ly, hoa)
    ds.append(ts)

# Lọc danh sách trúng tuyển và sắp xếp theo tổng điểm giảm dần
diem_chuan = 20
ds_trung_tuyen = [ts for ts in ds if ts.tong_diem() >= diem_chuan]
ds_trung_tuyen.sort(key=lambda x: x.tong_diem(), reverse=True)

print("\nDanh sách thí sinh trúng tuyển:")
for ts in ds_trung_tuyen:
    ts.in_thong_tin()
    print("-")
```

```python
class PS:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau
    def hop_le(self):
        return self.mau != 0
    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while self.mau == 0:
            print("Mẫu số phải khác 0. Vui lòng nhập lại.")
            self.mau = int(input("Nhập mẫu số: "))
    def in_ps(self):
        print(f"Phân số: {self.tu}/{self.mau}")

# Ví dụ sử dụng
ps = PS()
ps.nhap()
if ps.hop_le():
    ps.in_ps()
else:
    print("Phân số không hợp lệ!")
```

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.data = []  # Mảng float cấp phát động (dùng list Python)
    def isEmpty(self):
        return len(self.data) == 0
    def isFull(self):
        return len(self.data) >= self.capacity
    def push(self, value):
        if self.isFull():
            print("Ngăn xếp đã đầy!")
        else:
            self.data.append(float(value))
    def pop(self):
        if self.isEmpty():
            print("Ngăn xếp đã trống!")
            return None
        else:
            return self.data.pop()
    def count(self):
        return len(self.data)
    def __del__(self):
        print("Đã huỷ ngăn xếp.")

# Ví dụ sử dụng
stack = Stack(3)
stack.push(1.1)
stack.push(2.2)
stack.push(3.3)
print("Số phần tử trên ngăn xếp:", stack.count())
stack.push(4.4)  # Sẽ báo đầy
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())  # Sẽ báo trống
print("Số phần tử trên ngăn xếp:", stack.count())
```

##### Bài 5:

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.data = []
    def isEmpty(self):
        return len(self.data) == 0
    def isFull(self):
        return len(self.data) >= self.capacity
    def push(self, value):
        if self.isFull():
            print("Ngăn xếp đã đầy!")
        else:
            self.data.append(float(value))
    def pop(self):
        if self.isEmpty():
            print("Ngăn xếp đã trống!")
            return None
        else:
            return self.data.pop()
    def count(self):
        return len(self.data)
    def __del__(self):
        print("Đã huỷ ngăn xếp.")

# Chương trình chính
n = int(input("Nhập sức chứa của ngăn xếp: "))
stack = Stack(n)
while True:
    print("\n1. Push\n2. Pop\n3. Đếm số phần tử\n4. Thoát")
    choice = input("Chọn chức năng: ")
    if choice == '1':
        value = float(input("Nhập giá trị cần push: "))
        stack.push(value)
    elif choice == '2':
        result = stack.pop()
        if result is not None:
            print("Giá trị lấy ra:", result)
    elif choice == '3':
        print("Số phần tử trên ngăn xếp:", stack.count())
    elif choice == '4':
        break
    else:
        print("Chọn không hợp lệ!")
```

##### Bài 6:

```python
class NganXep:
    def __init__(self, kich_thuoc):
        self.kich_thuoc = kich_thuoc           
        self.mang = [0.0] * kich_thuoc         
        self.dinh = -1                         

    def __del__(self):  
        print(" Ngăn xếp đã được hủy.")

    # Kiểm tra ngăn xếp rỗng
    def rong(self):
        return self.dinh == -1

    # Kiểm tra ngăn xếp đầy
    def day(self):
        return self.dinh == self.kich_thuoc - 1

    # Đưa phần tử vào ngăn xếp
    def them(self, gia_tri):
        if self.day():
            print(" Ngăn xếp đã đầy, không thể thêm!")
        else:
            self.dinh += 1
            self.mang[self.dinh] = float(gia_tri)
            print(f"Đã thêm {gia_tri} vào ngăn xếp.")

    # Lấy phần tử ra khỏi ngăn xếp
    def lay(self):
        if self.rong():
            print(" Ngăn xếp rỗng, không thể lấy!")
            return None
        else:
            gia_tri = self.mang[self.dinh]
            self.dinh -= 1
            print(f" Đã lấy {gia_tri} ra khỏi ngăn xếp.")
            return gia_tri

    # Đếm số phần tử hiện tại trong ngăn xếp
    def dem(self):
        return self.dinh + 1

    # Hàm in nội dung ngăn xếp (yêu cầu Bài 6)
    def in_ngan_xep(self):
        if self.rong():
            print(" Ngăn xếp hiện tại: RỖNG")
        else:
            print(" Nội dung ngăn xếp (từ đáy → đỉnh): ", end="")
            for i in range(self.dinh + 1):
                print(self.mang[i], end=" ")
            print()
```

##### Bài 7:

```python
class Date:
    def __init__(self, ngay=1, thang=1, nam=2000):
        self.ngay = ngay
        self.thang = thang
        self.nam = nam

    # Hàm kiểm tra năm nhuận
    def nam_nhuan(self, nam):
        return (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0)

    # Số ngày trong 1 tháng
    def so_ngay_trong_thang(self, thang, nam):
        if thang in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif thang in [4, 6, 9, 11]:
            return 30
        elif thang == 2:
            return 29 if self.nam_nhuan(nam) else 28
        else:
            return 0

    # In ngày ra màn hình
    def hien_thi(self):
        print(f"{self.ngay:02d}/{self.thang:02d}/{self.nam}")

    # Tính ngày tiếp theo
    def ngay_tiep(self):
        ngay_max = self.so_ngay_trong_thang(self.thang, self.nam)
        if self.ngay < ngay_max:
            self.ngay += 1
        else:
            self.ngay = 1
            if self.thang == 12:
                self.thang = 1
                self.nam += 1
            else:
                self.thang += 1
```

##### Bài 9:

```python
class DaGiac:
    def __init__(self, canh=[]):
        self.canh = canh   # danh sách các cạnh

    def tinh_chu_vi(self):
        return sum(self.canh)
    
class HinhBinhHanh(DaGiac):
    def __init__(self, day, canh_ben, chieu_cao):
        super().__init__([day, canh_ben, day, canh_ben])
        self.day = day
        self.canh_ben = canh_ben
        self.chieu_cao = chieu_cao

    def tinh_dien_tich(self):
        return self.day * self.chieu_cao
    

class HinhChuNhat(HinhBinhHanh):
    def __init__(self, chieu_dai, chieu_rong):
        super().__init__(chieu_dai, chieu_rong, chieu_rong)  # chiều cao = chiều rộng
        self.chieu_dai = chieu_dai
        self.chieu_rong = chieu_rong

    def tinh_dien_tich(self):
        return self.chieu_dai * self.chieu_rong


class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        super().__init__(canh, canh)
        self.canh = canh

    def tinh_dien_tich(self):
        return self.canh * self.canh
```

```python
#Chương trình chính
if __name__ == "__main__":
    print("=== TÍNH TOÁN ĐA GIÁC ===")

    # Hình bình hành
    print("\n--- HÌNH BÌNH HÀNH ---")
    day = float(input("Nhập độ dài đáy: "))
    canh_ben = float(input("Nhập cạnh bên: "))
    chieu_cao = float(input("Nhập chiều cao: "))
    hbh = HinhBinhHanh(day, canh_ben, chieu_cao)
    print(f"Chu vi: {hbh.tinh_chu_vi()}")
    print(f"Diện tích: {hbh.tinh_dien_tich()}")

    # Hình chữ nhật
    print("\n--- HÌNH CHỮ NHẬT ---")
    cd = float(input("Nhập chiều dài: "))
    cr = float(input("Nhập chiều rộng: "))
    hcn = HinhChuNhat(cd, cr)
    print(f"Chu vi: {hcn.tinh_chu_vi()}")
    print(f"Diện tích: {hcn.tinh_dien_tich()}")

    # Hình vuông
    print("\n--- HÌNH VUÔNG ---")
    canh = float(input("Nhập cạnh hình vuông: "))
    hv = HinhVuong(canh)
    print(f"Diện tích: {hv.tinh_dien_tich()}")
```

##### Bài 10:

```python
import math

class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def hien_thi(self):
        return f"({self.x}, {self.y})"


class Elip(Diem):
    def __init__(self, x, y, ban_truc_lon, ban_truc_nho):
        super().__init__(x, y)  # tâm elip là 1 điểm
        self.a = ban_truc_lon   # bán trục lớn
        self.b = ban_truc_nho   # bán trục nhỏ

    def tinh_dien_tich(self):
        return math.pi * self.a * self.b

    def hien_thi(self):
        return (f"Tâm elip: {super().hien_thi()}, "
                f"bán trục lớn = {self.a}, bán trục nhỏ = {self.b}, "
                f"diện tích = {self.tinh_dien_tich():.2f}")


class DuongTron(Elip):
    def __init__(self, x, y, ban_kinh):
        super().__init__(x, y, ban_kinh, ban_kinh)  # a = b = r
        self.r = ban_kinh

    def tinh_dien_tich(self):
        return math.pi * self.r * self.r

    def hien_thi(self):
        return (f"Tâm đường tròn: {super().hien_thi()}, "
                f"bán kính = {self.r}, "
                f"diện tích = {self.tinh_dien_tich():.2f}")
```

```python
# Chương trình chính
if __name__ == "__main__":
    print("=== TÍNH TOÁN ELIP VÀ ĐƯỜNG TRÒN ===")

    # Nhập elip
    print("\n--- ELIP ---")
    x = float(input("Nhập hoành độ tâm: "))
    y = float(input("Nhập tung độ tâm: "))
    a = float(input("Nhập bán trục lớn a: "))
    b = float(input("Nhập bán trục nhỏ b: "))
    elip = Elip(x, y, a, b)
    print(elip.hien_thi())

    # Nhập đường tròn
    print("\n--- ĐƯỜNG TRÒN ---")
    x = float(input("Nhập hoành độ tâm: "))
    y = float(input("Nhập tung độ tâm: "))
    r = float(input("Nhập bán kính r: "))
    tron = DuongTron(x, y, r)
    print(tron.hien_thi())
```

##### Bài 11:

```python
import math

class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def kiem_tra_hop_le(self):
        return (self.a + self.b > self.c and
                self.a + self.c > self.b and
                self.b + self.c > self.a)

    def tinh_chu_vi(self):
        return self.a + self.b + self.c

    def tinh_dien_tich(self):
        p = self.tinh_chu_vi() / 2  # nửa chu vi
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

    def hien_thi(self):
        if not self.kiem_tra_hop_le():
            return "Tam giác không hợp lệ!"
        return (f"Tam giác có cạnh a={self.a}, b={self.b}, c={self.c}, "
                f"chu vi={self.tinh_chu_vi()}, "
                f"diện tích={self.tinh_dien_tich():.2f}")

class TamGiacVuong(TamGiac):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)

    def kiem_tra_vuong(self):
        canh = sorted([self.a, self.b, self.c])
        return math.isclose(canh[0]**2 + canh[1]**2, canh[2]**2)

    def hien_thi(self):
        if not self.kiem_tra_hop_le():
            return "Tam giác không hợp lệ!"
        if self.kiem_tra_vuong():
            return f"Tam giác vuông: {super().hien_thi()}"
        else:
            return "Không phải tam giác vuông!"

class TamGiacCan(TamGiac):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)

    def kiem_tra_can(self):
        return self.a == self.b or self.a == self.c or self.b == self.c

    def hien_thi(self):
        if not self.kiem_tra_hop_le():
            return "Tam giác không hợp lệ!"
        if self.kiem_tra_can():
            return f"Tam giác cân: {super().hien_thi()}"
        else:
            return "Không phải tam giác cân!"

class TamGiacDeu(TamGiac):
    def __init__(self, a):
        super().__init__(a, a, a)

    def hien_thi(self):
        return f"Tam giác đều: {super().hien_thi()}"
```

```python
#Chuơng trình chính
if __name__ == "__main__":
    print("=== KIỂM TRA TAM GIÁC ===")

    # Tam giác thường
    tg = TamGiac(3, 4, 5)
    print(tg.hien_thi())

    # Tam giác vuông
    tgv = TamGiacVuong(3, 4, 5)
    print(tgv.hien_thi())

    # Tam giác cân
    tgc = TamGiacCan(4, 4, 6)
    print(tgc.hien_thi())

    # Tam giác đều
    tgd = TamGiacDeu(5)
    print(tgd.hien_thi())
```
