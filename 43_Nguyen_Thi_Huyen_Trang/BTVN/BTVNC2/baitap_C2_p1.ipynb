# Câu hỏi

Câu 1: Khái niệm XML:
- XML (Extensible Markup Language) là ngôn ngữ đánh dấu mở rộng.
- Được tạo ra với mục đích phát triển và mở rộng các ngôn ngữ đánh dấu khác ngoài định dạng .txt, ASCII,…
- XML là ngôn ngữ đánh dấu được dùng để miêu tả dữ liệu.

Câu 2: Khái niệm DOM:
- DOM là tên viết tắt của (Document Object Model – có thể hiểu là Mô hình đối tượng tài liệu) là một chuẩn giao diện lập trình (API) độc lập nền tảng và độc lập ngôn ngữ, cho phép các chương trình, và các ngôn nghữ kịch bản (vd: JavaScrip, Python, PHP,.v.v.) truy cập và cập nhật động nội dung, cấu trúc của tài liệu HTML/XML.

Câu 3: Cú pháp tài liệu XML:
    <nodename>nội dung</nodename>
Trong đó:
- <nodename> là thẻ mở, tên của thẻ này do người dùng tự định nghĩa.
- </nodename> là thẻ đóng, tên của thẻ này phải trùng với tên của thẻ mở.
- content là nội dung của thẻ này

Câu 4: API XML  là là giao diện lập trình ứng dụng cho phép trao đổi dữ liệu ở dạng XML giữa các chương trình, hệ thống hoặc dịch vụ.

Bài 2.1:

```python
from xml.etree.ElementTree import Element, SubElement, tostring
from xml.dom import minidom

root = Element("Company")

name = SubElement(root, "Name")
name.text = "Công ty TNHH ABC"

director = SubElement(root, "Director")
fullname = SubElement(director, "FullName")
fullname.text = "Nguyễn Văn A"
title = SubElement(director, "Title")
title.text = "Giám đốc"

contact = SubElement(root, "Contact")
address = SubElement(contact, "Address")
address.text = "123 Đường lê lợi, Đà Nẵng"
phone = SubElement(contact, "Phone")
phone.text = "0123456789"

taxid = SubElement(root, "TaxID")
taxid.text = "0312345678"

# Chuyển cây XML thành chuỗi có định dạng đẹp
xml_str = minidom.parseString(tostring(root, "utf-8")).toprettyxml(indent="  ", encoding="utf-8")

# Ghi ra file company.xml
with open("company.xml", "wb") as f: 
    f.write(xml_str)

print(" Đã tạo xong file company")
```

```python
import xml.etree.ElementTree as ET 
from xml.dom import minidom

goc = ET.Element("DanhSachSinhVien")

# Thêm sinh viên 
sv1 = ET.SubElement(goc, "SinhVien")

ma1 = ET.SubElement(sv1, "MaSinhVien")
ma1.text = "24174600171"

ten1 = ET.SubElement(sv1, "TenSinhVien")
ten1.text = "Nguyễn Thị Huyền Trang"

namsinh1 = ET.SubElement(sv1, "NamSinh")
namsinh1.text = "2006"

lop1 = ET.SubElement(sv1, "Lop")
lop1.text = "DHKL18A3HN"

gioitinh1 = ET.SubElement(sv1, "GioiTinh")
gioitinh1.text = "Nữ"


sv2 = ET.SubElement(goc, "SinhVien")

ma2 = ET.SubElement(sv2, "MaSinhVien")
ma2.text = "24174600130"

ten2 = ET.SubElement(sv2, "TenSinhVien")
ten2.text = "Nguyễn Văn A"

namsinh2 = ET.SubElement(sv2, "NamSinh")
namsinh2.text = "2006"

lop2 = ET.SubElement(sv2, "Lop")
lop2.text = "DHKL18A3HN"

gioitinh2 = ET.SubElement(sv2, "GioiTinh")
gioitinh2.text = "Nam"

chuoi_xml = minidom.parseString(ET.tostring(goc, "utf-8")).toprettyxml(indent="  ", encoding="utf-8")

with open("danhsachsinhvien.xml", "wb") as f:
    f.write(chuoi_xml)

print(" Đã tạo xong file")
```

```python
doc = minidom.parse('sample.xml')
```

```python
goc = doc.documentElement
print("Tên phần tử gốc:", goc.tagName)

print("\nCác phần tử con của phần tử gốc:")
for node in goc.childNodes:
    if node.nodeType == node.ELEMENT_NODE:
        print("-", node.tagName)
```

Bài 2.5:

```python
from xml.dom import minidom

doc = minidom.parse('sample.xml')

danh_sach_phan_tu = doc.getElementsByTagName('*')  # '*' lấy tất cả các phần tử

print("Danh sách tên các phần tử:")
for phan_tu in danh_sach_phan_tu:
    print(phan_tu.tagName)
```
