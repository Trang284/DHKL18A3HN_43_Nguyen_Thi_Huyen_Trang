```python
import threading

# Hàm mà mỗi luồng sẽ chạy
def in_ten_luong():
    print("Tên luồng đang chạy:", threading.current_thread().name)

# Tạo danh sách luồng
threads = []

for i in range(5):
    t = threading.Thread(target=in_ten_luong, name=f"Luong-{i+1}")
    threads.append(t)
    t.start()

# Chờ tất cả luồng chạy xong
for t in threads:
    t.join()

print("Chương trình kết thúc")
```

Bài 6.2: 

```python
import threading
import urllib.request

# Hàm tải file
def tai_file(url, ten_file):
    print(f"Đang tải: {ten_file}")
    urllib.request.urlretrieve(url, ten_file)
    print(f"Tải xong: {ten_file}")

# Danh sách các file cần tải
files = [
    ("https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf", "file1.pdf"),
    ("https://www.learningcontainer.com/wp-content/uploads/2020/05/sample-text-file.txt", "file2.txt"),
    ("https://www.learningcontainer.com/wp-content/uploads/2020/04/sample-img.png", "file3.png")
]

threads = []

# Tạo và chạy các luồng
for url, ten in files:
    t = threading.Thread(target=tai_file, args=(url, ten))
    threads.append(t)
    t.start()

# Chờ các luồng hoàn thành
for t in threads:
    t.join()

print("Đã tải xong tất cả các tệp")
```

Bài 6.3: 

```python
import threading

# Luồng in số chẵn
def in_so_chan():
    print("Các số chẵn:")
    for i in range(30, 51):
        if i % 2 == 0:
            print(i, end=" ")
    print()

# Luồng in số lẻ
def in_so_le():
    print("Các số lẻ:")
    for i in range(30, 51):
        if i % 2 != 0:
            print(i, end=" ")
    print()

# Tạo 2 luồng
t1 = threading.Thread(target=in_so_chan)
t2 = threading.Thread(target=in_so_le)

# Chạy luồng
t1.start()
t2.start()

# Chờ 2 luồng hoàn thành
t1.join()
t2.join()

print("Chương trình kết thúc")
```

Bài 6.4: 

```python
import threading

# Danh sách lưu kết quả của các luồng
ket_qua = []

# Hàm tính tích trong một đoạn
def tinh_doan(start, end):
    tich = 1
    for i in range(start, end + 1):
        tich *= i
    ket_qua.append(tich)

n = 6   # số cần tính giai thừa

# Tạo 3 luồng
t1 = threading.Thread(target=tinh_doan, args=(1, 2))
t2 = threading.Thread(target=tinh_doan, args=(3, 4))
t3 = threading.Thread(target=tinh_doan, args=(5, 6))

# Chạy các luồng
t1.start()
t2.start()
t3.start()

# Chờ các luồng hoàn thành
t1.join()
t2.join()
t3.join()

# Nhân kết quả từ các luồng
giai_thua = 1
for k in ket_qua:
    giai_thua *= k

print(f"{n}! =", giai_thua)
```

```python
import threading
import requests

# Hàm gửi yêu cầu HTTP
def gui_yeu_cau(url):
    print(f"Đang gửi yêu cầu tới: {url}")
    response = requests.get(url)
    print(f"Hoàn thành: {url} | Mã trạng thái: {response.status_code}")

# Danh sách các URL
urls = [
    "https://www.example.com",
    "https://www.python.org",
    "https://www.wikipedia.org"
]

threads = []

# Tạo và chạy các luồng
for url in urls:
    t = threading.Thread(target=gui_yeu_cau, args=(url,))
    threads.append(t)
    t.start()

# Chờ các luồng hoàn thành
for t in threads:
    t.join()

print("Đã hoàn thành tất cả các yêu cầu HTTP")
```

```python
import threading
import time

N = 20  

def ghi_so_chan():
    print("Các số chẵn:")
    for i in range(0, N + 1, 2):
        print(i, end=" ")
        time.sleep(0.1)
    print()

def ghi_so_le():
    print("Các số lẻ:")
    for i in range(1, N + 1, 2):
        print(i, end=" ")
        time.sleep(0.1)
    print()

t1 = threading.Thread(target=ghi_so_chan)
t2 = threading.Thread(target=ghi_so_le)

t1.start()
t2.start()

t1.join()
t2.join()

print("Kết thúc chương trình")
```

Bài 6.8: 

```python
import threading
import random

def tinh_tong_list_con(list_con, index, ket_qua):
    tong = sum(list_con)
    print(f"Thread {index}: {list_con} -> Tổng = {tong}")
    ket_qua[index] = tong

n = int(input("Nhập số phần tử: "))
so_thread = int(input("Nhập số thread: "))

lst = [random.randint(0, 10) for _ in range(n)]
print("List ban đầu:", lst)

kich_thuoc = n // so_thread
threads = []
ket_qua = [0] * so_thread

bat_dau = 0

for i in range(so_thread):
    if i == so_thread - 1:
        list_con = lst[bat_dau:]
    else:
        list_con = lst[bat_dau: bat_dau + kich_thuoc]

    t = threading.Thread(
        target=tinh_tong_list_con,
        args=(list_con, i + 1, ket_qua)
    )
    threads.append(t)
    t.start()

    bat_dau += kich_thuoc

for t in threads:
    t.join()

tong_list = sum(ket_qua)
print("Tổng list =", tong_list)
```
