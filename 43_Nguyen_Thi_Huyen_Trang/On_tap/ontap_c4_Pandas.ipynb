##### Câu 1:
Pandas là thư viện Python dùng để xử lý, phân tích và thao tác dữ liệu dạng bảng (tabular data) một cách hiệu quả, đặc biệt với dữ liệu có nhãn (label).

##### Câu 2:
*Series
*DataFrame

##### Câu 3: Series là mảng một chiều có:

* Dữ liệu
* Index (nhãn)
* Kiểu dữ liệu

##### Câu 4: DataFrame là bảng dữ liệu hai chiều gồm:

* Hàng (rows)
* Cột (columns)
* Mỗi cột là một Series

##### Câu 5:
Index là tập nhãn dùng để định danh các hàng (hoặc cột) trong Series/DataFrame, giúp truy cập dữ liệu nhanh và chính xác.

##### Câu 6:
* Index: nhãn của hàng
* Column: nhãn của cột
* Index định vị theo trục 0, column theo trục 1
