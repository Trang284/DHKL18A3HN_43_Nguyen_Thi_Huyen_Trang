##### Khai báo lớp SinhVien có các thuộc tính masv, hoten, lop
* Viết phương thức info(), trả về thông tin của sinh viên
##### Định nghĩa lớp svk18a3 kế thừa từ lớp SinhVien
* Ghi đè phương thức info() 
##### Khai báo 1 đối tượng của lớp svk18a3 và gọi phương thức ghi đè để hiểu bản chất của đa hình.

```python
class SinhVien:
    def __init__(self, masv, hoten, lop):
        self.masv = masv
        self.hoten = hoten
        self.lop = lop

    def info(self):
        return f"Mã SV: {self.masv}, Họ tên: {self.hoten}, Lớp: {self.lop}"
```

```python
class svk18a3(SinhVien):
    def info(self):
        return f"[SV K18A3] {self.hoten} - MSSV: {self.masv} - Lớp: {self.lop}"
```

```python
sv = svk18a3("SV001", "Nguyễn Văn A", "K18A3")

print(sv.info())
```

Câu 12. Khi class con và cha có phương thức cùng tên, Python ưu tiên?
A. Phương thức class cha
B. Phương thức class con
C. Phương thức global
D. Phương thức static

```python
class A:
    def __init__(self):
        self.x = 10
a = A() #a là một đối tượng của lớp A?
print(a.x)
```

```python
class A:
    x = 5 # x=5 là thuộc tính j?
a = A()
print(a.x)
```

```python
class A:
    x = 5
a = A()
a.x = 10
print(A.x, a.x)
```

```python
class A:
    def show(self):
        print("A")
class B(A):
    pass
b = B() #vì B kế thừa A nên b có thể gọi phương thức show() của A
b.show()
```

```python
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B")
b = B()
b.show()
```
