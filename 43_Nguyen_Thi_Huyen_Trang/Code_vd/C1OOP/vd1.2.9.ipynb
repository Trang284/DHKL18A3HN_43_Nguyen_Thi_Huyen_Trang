```python
class Vector2D:
    # Biểu diễn một vector 2 chiều.
    def __init__(self, x: float = 0.0, y: float = 0.0):
        """Khởi tạo một đối tượng Vector2D mới.
          Tham số:
            x (float): Tọa độ x. Mặc định là 0.0.
            y (float): Tọa độ y. Mặc định là 0.0.
        """
        self.x = float(x)  # Tạo thuộc tính CỦA ĐỐI TƯỢNG ngay khi khởi tạo
        self.y = float(y)

    def set(self, x: float, y: float) -> None:  # tên hàm chữ thường theo PEP 8 - quy chuẩn viết mã Python (Python Enhancement Proposal #8)
        self.x = float(x)
        self.y = float(y)

def main() -> None:
    vec = Vector2D()  # x=0.0, y=0.0 của CHÍNH đối tượng vec
    vec.set(5, 6)
    print(f"X: {vec.x}, Y: {vec.y}")  # X: 5.0, Y: 6.0

if __name__ == "__main__":
    main()
```
