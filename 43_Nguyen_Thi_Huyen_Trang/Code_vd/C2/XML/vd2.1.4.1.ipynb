```python
import xml.etree.ElementTree as ET # Sử dụng thư viện ElementTree để làm việc với XML Bước 1

tree = ET.parse('employee.xml') # Sử dụng phương thức parse() để tải và phân tích cú pháp một tệp XML Bước 2

root = tree.getroot() 
name = root.find('name') #Tìm kiếm thẻ name Bước 3
print(name.text)

expertise = root.find('expertise') #Tìm kiếm thẻ expertise Bước 4
print(expertise.attrib['name'])

city = root.find('city') #Tìm kiếm thẻ city Bước 5
print(city.text)

newexpertise = ET.Element('expertise') #Tạo thẻ expertise mới Bước 6
newexpertise.attrib['name'] = 'XML'    #Gán thuộc tính cho thẻ expertise
root.append(newexpertise)

tree.write('updated.xml') #Ghi lại thay đổi vào 1 tệp XML mới Bước 7
```
